```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
brussard=pd.read_csv("brussard.csv")
brussardcontrol=brussard.iloc[12:24]
brussardvirus=brussard.iloc[24:40]
brussardinfected=brussard.iloc[0:12]

fig,ax=plt.subplots(figsize=(10,5))
plt.title('Relationship between Micromonas Pusilla and Virus MpV')
ax2=ax.twinx()
ax.plot(brussardcontrol.time,brussardcontrol.abundance,"-r",label="Non-Infected")
ax.plot(brussardinfected.time,brussardinfected.abundance,"-y",label="Infected")
ax2.plot(brussardvirus.time,brussardvirus.abundance,"-b",label="MpV")
ax.legend()
ax2.legend(loc='upper right')
ax.set_xlabel('time(h)')
ax.set_ylabel('Host Abundance (mL)')
ax2.set_ylabel('Viral Abundance (mL)')
```

```python
brussardPp=pd.read_csv("brussard4bc.csv")
ppearlyinfected=brussardPp.iloc[0:10] #extract earlyphase infected
ppearlycontrol=brussardPp.iloc[10:20]
ppearlyvirus=brussardPp.iloc[20:31]

fig,ax=plt.subplots(figsize=(10,5))
plt.title('Relationship between Phaeocystis Pouchetii and Virus PpV (Early phase)')
ax2=ax.twinx()
ax.plot(ppearlycontrol.time,ppearlycontrol.abundance,"-r",label="Non-Infected")
ax.plot(ppearlyinfected.time,ppearlyinfected.abundance,"-y",label="Infected")
ax2.plot(ppearlyvirus.time,ppearlyvirus.abundance,"-b",label="PpV")
ax.legend()
ax2.legend(loc='upper right')
ax.set_xlabel('time(h)')
ax.set_ylabel('Host Abundance (mL)')
ax2.set_ylabel('Viral Abundance (mL)')
```

```python
pplateinfected=brussardPp.iloc[31:44] #extract infected data
pplatecontrol=brussardPp.iloc[44:57] #extract non-infected host
pplatevirus=brussardPp.iloc[57:71] #extract virus data
fig,ax=plt.subplots(figsize=(10,5))
plt.title('Relationship between Phaeocystis Pouchetii and Virus PpV (late phase)')
ax2=ax.twinx()
ax.plot(pplatecontrol.time,pplatecontrol.abundance,"-r",label="Non-Infected")
ax.plot(pplateinfected.time,pplateinfected.abundance,"-y",label="Infected")
ax2.plot(pplatevirus.time,pplatevirus.abundance,"-b",label="PpV")
ax.legend(loc='upper left')
ax2.legend(loc='upper right')
ax.set_xlabel('time(h)')
ax.set_ylabel('Host Abundance (mL)')
ax2.set_ylabel('Viral Abundance (mL)')
```

```python
m_polaris3=pd.read_csv("Maat3polaris.csv") 
mpolariscontrol=m_polaris3.iloc[0:10] #m.polaris control
mpolarisinfected44=m_polaris3.iloc[10:19] #m.polaris infected iwht 44-T
mpolarisvirus44=m_polaris3.iloc[19:31] #virus 44-T
mpolarisinfected45=m_polaris3.iloc[31:37] #m.polaris infected with 45-T
mpolarisvirus45=m_polaris3.iloc[37:46] #virust 45-T
mpolarisinfected46=m_polaris3.iloc[46:52] #m.polaris infected with 46-T
mpolarisvirus46=m_polaris3.iloc[52:61] #virus 46-T
mpolarisinfected47=m_polaris3.iloc[61:67] #m.polaris infected with 47-T
mpolarisvirus47=m_polaris3.iloc[67:75] #virus 47-T

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Abundance of Micromonas Polaris')
plt.plot(mpolarisinfected44.time,mpolarisinfected44.abundance,"-or",label="44-T")
plt.plot(mpolarisinfected45.time,mpolarisinfected45.abundance,"-oy",label="45-T")
plt.plot(mpolarisinfected46.time,mpolarisinfected46.abundance,"-ob",label="46-T")
plt.plot(mpolarisinfected47.time,mpolarisinfected47.abundance,"-og",label="47-T")
plt.plot(mpolariscontrol.time,mpolariscontrol.abundance,"-oc",label="Control")
plt.legend()
plt.xlabel('time post infection (h)')
plt.ylabel('Host Abundance (mL)')

plt.subplot(1,2,2)
plt.title('Abundance of MpV')
plt.plot(mpolarisvirus44.time,mpolarisvirus44.abundance,"-or",label="44-T")
plt.plot(mpolarisvirus45.time,mpolarisvirus45.abundance,"-oy",label="45-T")
plt.plot(mpolarisvirus46.time,mpolarisvirus46.abundance,"-ob",label="46-T")
plt.plot(mpolarisvirus47.time,mpolarisvirus47.abundance,"-og",label="47-T")
plt.legend()
plt.xlabel('time post infection (h)')
plt.ylabel('Viral Abundance (mL)')
```
